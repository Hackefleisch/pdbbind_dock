This file is intended to extract all docking information as tables from a h5 directory

```python
from tqdm import tqdm
from data import PDBRinterface
import pandas as pd

target = 'pdbbind'

results = PDBRinterface(target + '_h5/')
print(target, len(results.results), 'results were found')

df_list = []

for pdb in tqdm(results.results):
    try:
        res = results.get_result(pdb, False)
    except:
        continue
    
    if len(res.docking_df[res.docking_df.type == 'apo_relax_docking_std']) != 300:
        continue

    df = res.docking_df

    df.insert(0, 'pdb', pdb)


    df_list.append(df)

print(len(df_list), 'finished completely and were added to list.')

combined_df = pd.concat(df_list)
combined_df.to_csv(target + '_docking_data.zip')
```
