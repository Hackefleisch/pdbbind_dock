```python
from pyrosetta import *

pyrosetta.init(options='-in:auto_setup_metals -ex1 -ex2 -restore_pre_talaris_2013_behavior true -out:level 300', silent=True)

xml_objects = rosetta.protocols.rosetta_scripts.XmlObjects.create_from_file('xml_protocols/docking_std.xml')
scfx = xml_objects.get_score_function("hard_rep")
scfx.set_weight( rosetta.core.scoring.coordinate_constraint, 0.5 )
```

```python
from rdkit import Chem
from rdkit.Geometry import Point3D
import copy

mol = copy.deepcopy(results['conformers'])
with Chem.SDWriter('orig.sdf') as w:
    w.SetKekulize(False)
    for i in range(mol.GetNumConformers()):
        w.write(mol, i)
```

```python
ref_mol = copy.deepcopy(results['conformers'])
for i in range(1, ref_mol.GetNumConformers()):
    ref_mol.RemoveConformer(i)

mols = []

docking_protocol = 'docking_perturb_pose'

for i in range(len(results['docking_results'][docking_protocol])):
    res = results['docking_results'][docking_protocol][i]
    print(res['total_score'])
    print(res['idelta_score'])
    base_pdb_strarr = results['input_complexes'][res['input_pose_name']]['pdb_string_arr']
    for idx, upd in res['pdb_string_delta'].items():
        base_pdb_strarr[idx] = upd
    for line in base_pdb_strarr:
        if line[:6] == 'HETATM' and line[17:20] == 'UNK':
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            name = line[12:16]
            ref_mol.GetConformer().SetAtomPosition(results['atm_to_idx'][name], Point3D(x,y,z))
    mols.append(copy.deepcopy(ref_mol))

print(len(results['docking_results'][docking_protocol]))
print(len(mols))

with Chem.SDWriter('dock.sdf') as w:
    w.SetKekulize(False)
    for m in mols:
        w.write(m)
```

```python
import load_ligand

with open('run.pickle', 'rb') as file:
    results = pickle.load(file)

for iiii in range(len(results['docking_results']['docking_std_pose'])):

    res = results['docking_results']['docking_std_pose'][iiii]

    base_pdb_strarr = results['input_complexes'][res['input_pose_name']]['pdb_string_arr']
    for idx, upd in res['pdb_string_delta'].items():
        base_pdb_strarr[idx] = upd

    pdb_string = []
    for line in base_pdb_strarr:
        pdb_string.append(line)
        if line[:3] == "TER":
            break
    pdb_string = "\n".join(pdb_string)

    # just to manually check the file and watch it in pymol
    with open('test.pdb', 'w') as file:
        file.write(pdb_string)

    pose = rosetta.core.pose.Pose()
    rosetta.core.import_pose.pose_from_pdbstring(pose, pdb_string)

    mol = copy.deepcopy(results['conformers'])
    for i in range(1, ref_mol.GetNumConformers()):
        ref_mol.RemoveConformer(i)
    for line in base_pdb_strarr:
        if line[:6] == 'HETATM' and line[17:20] == 'UNK':
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            name = line[12:16]
            mol.GetConformer().SetAtomPosition(results['atm_to_idx'][name], Point3D(x,y,z))

    mut_res, index_to_vd = load_ligand.rdkit_to_mutable_res(mol)
    residue = load_ligand.mutable_res_to_res(mut_res)

    pose.append_residue_by_jump( residue, 1, "", "", True )
    pose.pdb_info().chain( pose.total_residue(), 'X' )
    pose.update_pose_chains_from_pdb_chains()

    pose.dump_pdb("test2.pdb")

    print(scfx(pose), res['total_score'], scfx(pose)- res['total_score'])
```
